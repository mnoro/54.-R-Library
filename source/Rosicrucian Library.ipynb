```python
import io

from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfpage import PDFPage

def extract_text_from_pdf(pdf_path):
    resource_manager = PDFResourceManager()
    fake_file_handle = io.StringIO()
    converter = TextConverter(resource_manager, fake_file_handle)
    page_interpreter = PDFPageInterpreter(resource_manager, converter)

    with open(pdf_path, 'rb') as fh:
        for page in PDFPage.get_pages(fh, 
                                      caching=True,
                                      check_extractable=True):
            page_interpreter.process_page(page)

        text = fake_file_handle.getvalue()

    # close open handles
    converter.close()
    fake_file_handle.close()

    if text:
        return text
```

# Process documents including uploading to ElasticSearch

```python
pdfs = sorted(fnmatch.filter(os.listdir(pathIn0), '*.pdf'))
pdfs
```

```python
inventory = pd.DataFrame()
idx = 0
res = 2000

# Image zoom factor
zoom = 5
mat = fitz.Matrix(zoom, zoom)
```

```python
portNos = ['9210']
```

```python
for portNo in portNos:
    try:
        print('Port Number %s'%(portNo))
        client = Elasticsearch("http://localhost:%s"%(portNo))
        elastic_info = Elasticsearch.info(client)
        print("Cluster Info:", json.dumps(elastic_info, indent=4))
    except Exception as err:
        print('Elasticsearch client Error:',err)
        client = None
```

```python
es = Elasticsearch(['http://localhost:%s'%(portNos[0])])
```

```python
indexName = 'maxlibrary'
```

```python
cols = ['Title', 'FullPath', 'NumPages', 'DocumentInfo',
        'ProcessingTime','Content', 'ProcessingTimeOCR(ms)']
```

```python
for f in pdfs:
    fn = os.path.join(pathIn0, f)
    pdf_file = fitz.open(fn)
    print(fn)
    # Iterate over pdf pages
    tmpdf = pd.DataFrame()
    tot_images = 0
    tmpdf.at[idx, 'Title']=f
    tmpdf.at[idx, 'FullPath']=fn
    tmpdf.at[idx,'NumPages']=len(pdf_file)
    tmpdf.at[idx,'DocumentInfo']=str(pdf_file.metadata)
    tmpdf.at[idx, 'ProcessingTime']=str(datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
    startOcr = datetime.datetime.now()
    # Add Author, DocumentInfo, AbsolutePath, ProcessingTime, ProcessingTimeOCR, EndOCR
    content = ''
    for page_index in range(len(pdf_file)):
        page = pdf_file[page_index]
        pix_page = page.getPixmap(matrix = mat)
        #image_list = page.getImageList()
        pix_image_name = os.path.join(pathOut, f"{f}-image{page_index+1}_pageAsImage.jpg")
        pix_page.writeImage(pix_image_name,'jpg')
        
        # OCR
        pageText = pytesseract.image_to_string(Image.open(pix_image_name))
        #pageText = pytesseract.image_to_string(pix_page)
        pageText = '<h4>Header Page: %s</h4>\n'%(page_index+1) + pageText +'<h4>Footer Page: %s</h4>\n'%(page_index+1)
        content += pageText
    tmpdf.at[idx,'Content']=content
    endOcr = datetime.datetime.now()
    tmpdf.at[idx, 'ProcessingTimeOCR(ms)']=(endOcr-startOcr)
    
    print('OCR took %s milliseconds for %s pages'%(endOcr-startOcr, len(pdf_file)))
    # Add to Elasticsearch
    # turn into json
    data = pd.Series([      
        tmpdf.at[idx,'Title'],
        tmpdf.at[idx,'FullPath'],
        tmpdf.at[idx,'NumPages'],
        tmpdf.at[idx,'DocumentInfo'],
        tmpdf.at[idx,'ProcessingTime'],
        tmpdf.at[idx,'Content'],
        tmpdf.at[idx,'ProcessingTimeOCR(ms)']], index = cols)
    data_json = data.to_json()
    json_data = json.loads(data_json)
    # Send to elastic
    res = es.index(index = indexName,  body = json_data)    
    
    # Process Next
    idx +=1
    inventory = inventory.append(tmpdf)
        
```
